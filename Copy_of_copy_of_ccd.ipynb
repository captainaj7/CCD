```python
#Taking Cloud Cover to be the target variable
#Copied the data in X after dropping the target variable
#Copied the data of target variable in Y
X = train.drop('Total Cloud Cover [%]' , axis=1)
Y = train['Total Cloud Cover [%]']

print(X.shape)
print(Y.shape)
```

SPLITTING THE DATA IN TEST AND TRAIN SPLIT

80% TRAINING DATA

20% TESTING DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
print('Shape of X_train = ', X_train.shape)
print('Shape of Y_train = ', Y_train.shape)
print('Shape of X_train = ', X_test.shape)
print('Shape of Y_train = ', Y_test.shape)
```

```python
print((X_train))
print((Y_train))
```

Random Forest Regression - ML Model Training

```python
from sklearn.ensemble import RandomForestRegressor
```

Creation of 100 Decision trees after training using RANDOM FOREST REGRESSOR ALGORITHM

```python
regressor = RandomForestRegressor(n_estimators=100, criterion='squared_error', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=X_train.shape[1], max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False  )
regressor.fit(X_train, Y_train)
```

```python
#To make predictions on the test data, we pass the X_test values to the predict() method. We can assign the results to the variable y_pred:

y_pred = regressor.predict(X_test)
y_pred
```

```python
#A constant model that always predicts the expected value of y
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```
